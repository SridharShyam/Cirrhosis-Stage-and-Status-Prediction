```python
missing_percent = data.isnull().mean() * 100
print("Percentage of Missing Values:\n", missing_percent)

missing_rows_count = data.isnull().any(axis=1).sum()
print(f"Rows with missing values: {missing_rows_count}")

total_rows = len(data)
removal_percent = (missing_rows_count / total_rows) * 100
print(f"Percentage of rows to be removed: {removal_percent:.2f}%")

data_clean = data.dropna()

print(f"Original Dataset Size: {data.shape}")
print(f"Cleaned Dataset Size: {data_clean.shape}")
```

```python
data=data.dropna(axis=0,how='any')
data
```

```python
data=data.drop(columns='ID')
data
```

```python
data['Sex'] = data['Sex'].map({'M': 0, 'F': 1})
data['Ascites'] = data['Ascites'].map({'N': 0, 'Y': 1})
data['Hepatomegaly']=data['Hepatomegaly'].map({'N':0,'Y':1})
data['Spiders']=data['Spiders'].map({'N':0,'Y':1})
```

```python
data
```

```python
data['Edema']=data['Edema'].map({'N':0,'Y':1,'S':2})
```

```python
data['Status']=data['Status'].map({'D':0,'C':1,'CL':2})
```

```python
data
```

```python
data=pd.get_dummies(data, columns=['Drug'])
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
data['Status']=le.fit_transform(data['Status'])
data
```

```python
from sklearn.model_selection import train_test_split

features=data.drop(columns=['Status','Stage'])

X_stage=features
Y_stage=data['Stage']

X_stage_train,X_stage_test,Y_stage_train,Y_stage_test=train_test_split(X_stage,Y_stage,test_size=0.2,random_state=50)

X_status=features
Y_status=data['Status']

X_status_train,X_status_test,Y_status_train,Y_status_test=train_test_split(X_status,Y_status,test_size=0.2,random_state=50)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report

RF=RandomForestClassifier()

RF.fit(X_stage_train,Y_stage_train)
stage_pred=RF.predict(X_stage_test)

print("Cirrhosis Stage Prediction:")
print(stage_pred)
```

```python
print("Classification Report:")
print(classification_report(Y_stage_test,stage_pred))
print("Accuracy Score:",accuracy_score(Y_stage_test,stage_pred)*100)
```

```python
RF.fit(X_status_train,Y_status_train)
status_pred=RF.predict(X_status_test)

print("Cirrhosis Status Prediction:")
print(status_pred)
```

```python
print("Classification Report:")
print(classification_report(Y_status_test,status_pred))
print("Accuracy Score:",accuracy_score(Y_status_test,status_pred)*100)
```
